```python
import pandas as pd
import re
import string
from bs4 import BeautifulSoup
import nltk
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
import spacy
```

```python
# Load spacy
nlp = spacy.load('en_core_web_sm')
nltk.download('stopwords')

def clean_string(text, stem="None"):

    final_string = ""

    # Make lower
    text = text.lower()

    # Remove line breaks
    # Note: that this line can be augmented and used over
    # to replace any characters with nothing or a space
    text = re.sub(r'\n', '', text)


    cleaned_text = re.sub(r'US \d+ [,\/\d ]* B\d+|\( \d+ \)|\( \d+ \d+ \)|\( *U *S *\) \d+,\d+,\d+ *A|\(C *ontinued \) .*? \(C *ontinued \)|\( *Notice:.*? \)|\( *C *ontinued \) \d{1,2} \/\d{1,2} \d{4}|\( *C *ontinued \)', '', text)
    
    
    # Remove remaining numerical characters and punctuation
    cleaned_text = re.sub(r'\d+|[^a-zA-Z\s]', '', cleaned_text)
    
    # Tokenize the text into individual words
    cleaned_text = cleaned_text.split()
    
    # Join the words back into a single string
    cleaned_text = ' '.join(cleaned_text)



    return cleaned_text
```

```python
import pandas as pd

df = pd.read_csv("raw_data/output.csv")
df.head(2)
```

```python
df = df.dropna(subset=["page_content"])
```

```python
df["cleaned_text"] = df["page_content"].apply(lambda x: clean_string(x))
```
